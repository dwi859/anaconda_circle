```python
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2

import numpy as np

theta = np.linspace(0,2 * np.pi, 100)

fig, ax = plt.subplots(1)

r = np.sqrt(20.0)
r2 = np.sqrt(18.0)
r3 = np.sqrt(15.0)
r4 = np.sqrt(13.0)



x1 = r*np.cos(theta)
x2 = r*np.sin(theta)
x3 = r2*np.cos(theta)
x4 = r2*np.sin(theta)
x5 = r3*np.cos(theta)
x6 = r3*np.sin(theta)
x7 = r4*np.cos(theta)
x8 = r4*np.sin(theta)



ax.plot(x1, x2 , x3, x4,'--r', x5, x6,'*g', x7, x8,'y')




plt.xlim(-4.95, 4.95, -2.95, 2.95)
plt.ylim(-4.95, 4.95, -2.95, 2.95)




plt.show()
```
